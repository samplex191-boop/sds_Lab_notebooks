```python
import random
import numpy as np
import matplotlib.pyplot as plt
```

## 1. Generate random integer number in a range 0 to 100 using Python.

```python
rand_int = np.random.randint(0, 100)
print("Random Integer:", rand_int)
```

## 2. Generate floating point number in a range 0 to 100 using Python.  

```python
rand_float = np.random.uniform(0, 100)
print("Random Float:", rand_float)
```

##3. Create a list of size 1000 random integer numbers in a range 0 to 100 and store in variable named as population using Python.

```python
population_int = np.random.randint(0, 101, 1000)
print("Integer population created.")
```

## 4. Construct a histogram of population created in Question3 using Python.

```python
plt.hist(population_int, bins=30, edgecolor='black')
plt.title("Histogram of Integer Population")
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.show()
```

##5. Find mean and variance of population created in Question3 using Python.

```python
mean_int = np.mean(population_int)
var_int = np.var(population_int)
print("Mean (Integer Population):", mean_int)
print("Variance (Integer Population):", var_int)
```

## 6. Draw a random sample of size n (user defined) from population created in Question3 using Python.

```python
n = int(input("Enter sample size : "))
sample_int = np.random.choice(population_int, size=n, replace=False)
print("Sample drawn from integer population:", sample_int[:10])
```

## 7. Create distribution of sample mean of samples drawn from population created in Question3 using Python. Here sample size and number of sample should be user defined.  

```python
num_samples = int(input("Enter number of samples: "))
sample_means_int = []

for _ in range(num_samples):
    sample = np.random.choice(population_int, size=n, replace=True)
    sample_means_int.append(np.mean(sample))

print(f"{num_samples} samples created.")
```

## 8. Construct a histogram of distribution of sample mean created in Question7 using Python.


```python
plt.hist(sample_means_int, bins=30, edgecolor='black')
plt.title("Sampling Distribution of Sample Mean (Integer Population)")
plt.xlabel("Sample Mean")
plt.ylabel("Frequency")
plt.show()
```

## 23. We have population data for individual smoking habits. We know that the true population proportion for smoking is 0.395 or 39.5%. For the given data verify Central Limit Theorem for Sampling distribution for Sample proportions using Python. Take population size equal to 1000.

```python
p = 0.395
population_smoking = np.random.binomial(1, p, 1000)

sample_size_prop = int(input("Enter sample size for proportion: "))
num_samples_prop = int(input("Enter number of samples for proportion: "))

sample_proportions = []
for _ in range(num_samples_prop):
    sample = np.random.choice(population_smoking, size=sample_size_prop, replace=True)
    sample_proportions.append(np.mean(sample))

plt.hist(sample_proportions, bins=30, density=True, edgecolor='black')
plt.title("CLT Verification for Sample Proportions (Smoking Data)")
plt.xlabel("Sample Proportion")
plt.ylabel("Density")
plt.show()

print("True Population Proportion:", p)
print("Mean of Sample Proportions:", np.mean(sample_proportions))
print("Conclusion: Sampling distribution of sample proportion is approximately normal (CLT verified).")
```
