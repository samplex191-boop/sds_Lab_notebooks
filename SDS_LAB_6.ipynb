```python
from scipy.stats import norm

p1 = norm.cdf(1.5)

p2 = 1 - norm.cdf(0.5)

p3 = norm.cdf(1.5) - norm.cdf(-1.5)

p4 = 1 - (norm.cdf(0.5)-norm.cdf(-0.5))

p5 = norm.cdf(1) - norm.cdf(-2.2)

print(f"P(Z<1.5)={p1}")
print(f"P(Z>0.5)={p2}")
print(f"P(|Z|<1.5)={p3}")
print(f"P(|Z|>0.5)={p4}")
print(f"P(-2.2<Z<1)={p5}")
```

```python
z1 = norm.ppf(.9)
z2 = norm.ppf(.95)
z3 = norm.ppf(.99)

z4 = norm.ppf((0.9+1)/2)
z5 = norm.ppf((0.95+1)/2)
z6 = norm.ppf((0.99+1)/2)


print(f"Z(0.9)={z1}")
print(f"Z(0.95)={z2}")
print(f"Z(0.99)={z3}")
print(f"|Z|(0.9)={z4}")
print(f"|Z|(0.95)={z5}")
print(f"|Z|(0.99)={z6}")
```

```python
from scipy.stats import t

p1 = t.cdf(1.5, 20)
p2 = 1 - t.cdf(0.5, 15)

p3 = t.cdf(1.5, 25) - t.cdf(-1.5, 25)
p4 = 1 - (t.cdf(0.5, 35)-t.cdf(-0.5, 35))

p5 = t.cdf(1, 42) - t.cdf(-2.2, 42)

print(f"P(t < 1.5) with d.o.f = 20 is {p1}")
print(f"P(t > 0.5) with d.o.f = 15 is {p1}")
print(f"P(|t| < 1.5) with d.o.f = 25 is {p1}")
print(f"P(|t| > 0.5) with d.o.f = 35 is {p1}")
print(f"P(-2.2 < t < 1) with d.o.f = 42 is {p1}")
```

```python
from scipy.stats import f

p1 = f.cdf(1.5, 5, 14)
p2 = 1 - f.cdf(2.5, 15, 14)
p3 = f.cdf(4.1, 13, 17) - f.cdf(0.5, 13, 17)

print(f"P(F < 1.5) with df1=5, df2=14 is {p1}")
print(f"P(F > 2.5) with df1=15, df2=14 is {p2}")
print(f"P(0.5 < F < 4.1) with df1=13, df2=17 is {p3}")
```

```python
F1 = f.ppf(0.9, 5, 14)
F2 = f.ppf(0.95, 15, 13)
F3 = f.ppf(0.99, 25, 28)

print(f"F(0.9, 5, 14)={F1}")
print(f"F(0.95, 15, 13)={F2}")
print(f"F(0.99, 25, 28)={F3}")
```

```python
from scipy.stats import chi2

p1 = chi2.cdf(1.5, 10)
p2 = 1 - chi2.cdf(2.5, 5)
p3 = chi2.cdf(4.1, 2) - chi2.cdf(0.5, 2)

print(f"P(X < 1.5) with df = 10 is {p1}")
print(f"P(X > 2.5) with df = 5 is {p2}")
print(f"P(0.5 < X < 4.1) with df = 2 is {p3}")
```

```python
X1 = chi2.ppf(0.9, 1)
X2 = chi2.ppf(0.95, 3)
X3 = chi2.ppf(0.99, 2)

X4 = chi2.ppf(0.05, 1)
X5 = chi2.ppf(0.025, 3)
X6 = chi2.ppf(0.005, 2)

print(f"X(0.9, 1) = {X1}")
print(f"X(0.95, 3) = {X2}")
print(f"X(0.99, 2) = {X3}")
print(f"X(0.05, 1) = {X4}")
print(f"X(0.025, 3) = {X5}")
print(f"X(0.005, 2) = {X6}")
```
