```python
from scipy.stats import binom

n = 800
p1 = binom.pmf(k=2, n=4, p=.5)
print(f"{n*p1} families have 2 cats and 2 dogs.")


p2 = binom.pmf(k=0, n=4, p=.5)
print(f"{n*(1-p2)} families have atleast 1 dog.")


p3 = binom.pmf(k=0, n=4, p=.5) + binom.pmf(k=1, n=4, p=.5) + binom.pmf(k=2, n=4, p=.5)
print(f"{n*p3} families have at most 2 cats.")


p4 = binom.pmf(k=1, n=4, p=.5) + binom.pmf(k=2, n=4, p=.5) + binom.pmf(k=3, n=4, p=.5)
print(f"{n*p4} families have both cat and dog.")
```

```python
from scipy.stats import poisson

l = 1.8

p1 = poisson.pmf(0, l)
print(f"probability of computer will function without a breakdown is {p1}")

p2 = poisson.pmf(1, l)
print(f"probability of computer will function with only 1 breakdown is {p2}")

p3 = 1 - (p1+p2)
print(f"probability of computer will function with atleast 2 breakdowns is {p3}")
```

```python
from scipy.stats import hypergeom

p1 = hypergeom.pmf(3, 20, 5, 12)
print(f"With 3 ambassadors {p1} ")

p2 = 1 - hypergeom.pmf(0, 20, 5, 12) - hypergeom.pmf(1, 20, 5, 12) - hypergeom.pmf(2, 20, 5, 12)
print(f"With atleast 3 ambassadors {p2}")

p3 = hypergeom.pmf(0, 20, 5, 12) + hypergeom.pmf(5, 20, 5, 12)
print(f"With 5 of the same make {p3}")
```

```python
from scipy.stats import uniform

p1 = uniform.cdf(5, 0, 15)
print(f"less than 5 minutes : {p1}")

p2 = 1-uniform.cdf(12, 0, 15)
print(f"atleast 12 minutes : {p2}")
```

```python
from scipy.stats import expon

t=40000

p1 = expon.cdf(20000,scale=t)
print(f"one of the tyres last atleast 20000km : {1 - p1}")

p2 = expon.cdf(30000,scale=t)
print(f"one of the tyres last atleast 30000km : {p2}")
```

```python
from scipy.stats import gamma

p =  gamma.cdf(12, 3, 0, 2)
print(f"Probability that power supply is inadequate on a given day is {1-p} ")
```

```python
from scipy.stats import beta

p = beta.cdf(0.3, 2, 5) -  beta.cdf(0.2, 2, 5)
print(f"Probability of defective balls from 20% to 30% is {p}")
```

```python
from scipy.stats import norm

p= 1 - norm.cdf(75, loc=65, scale=5)**3
print(f"Probability that atleast one student would have score above 75 is {p}")
```
